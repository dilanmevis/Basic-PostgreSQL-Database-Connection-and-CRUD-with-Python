```python
#K-Means Clustering Visualization of Age and Annual Income with 5 Clusters
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

data = pd.read_csv("/Users/mac/Desktop/Mall_Customers.csv")

data_selected = data[['Age', 'Annual Income (k$)']]

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_selected)

kmeans = KMeans(n_clusters=5, random_state=123)
kmeans_result = kmeans.fit_predict(data_scaled)

data['cluster'] = kmeans_result

plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Annual Income (k$)'], c=data['cluster'], cmap='viridis', s=50)
plt.title('K-Means Clustering (5 clusters)')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.colorbar(label='Cluster')
plt.show()
```

```python
#K-Means Clustering of Customers by Income and Spending Score
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("/Users/mac/Desktop/Mall_Customers.csv")

data_selected = data[['Annual Income (k$)', 'Spending Score (1-100)']]

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_selected)

kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(data_scaled)

fig, ax = plt.subplots(figsize=(14, 6))

ax.scatter(data_selected[y_kmeans == 0].iloc[:, 0], data_selected[y_kmeans == 0].iloc[:, 1], s=100, c='red', label='Cluster 1')
ax.scatter(data_selected[y_kmeans == 1].iloc[:, 0], data_selected[y_kmeans == 1].iloc[:, 1], s=100, c='blue', label='Cluster 2')
ax.scatter(data_selected[y_kmeans == 2].iloc[:, 0], data_selected[y_kmeans == 2].iloc[:, 1], s=100, c='green', label='Cluster 3')
ax.scatter(data_selected[y_kmeans == 3].iloc[:, 0], data_selected[y_kmeans == 3].iloc[:, 1], s=100, c='cyan', label='Cluster 4')
ax.scatter(data_selected[y_kmeans == 4].iloc[:, 0], data_selected[y_kmeans == 4].iloc[:, 1], s=100, c='magenta', label='Cluster 5')


plt.title('Cluster Segmentation of Customers')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
# DBSCAN
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

data = pd.read_csv("/Users/mac/Desktop/Mall_Customers.csv")

data_selected = data[['Age', 'Annual Income (k$)']]

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_selected)

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_result = dbscan.fit_predict(data_scaled)

data['dbscan_cluster'] = dbscan_result

plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Annual Income (k$)'], c=data['dbscan_cluster'], cmap='viridis', s=50)
plt.title('DBSCAN Clustering')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.colorbar(label='Cluster')
plt.show()
```



The graph shows customer clustering based on **Age** and **Annual Income** features using the DBSCAN (Density-Based Spatial Clustering of Applications with Noise) algorithm. Based on the visualization, the following observations can be made:

---

### **1. Number of Clusters:**

* As seen from the colors, **only 1 cluster** is detected (`cluster 0`, yellow color).
* All other points are considered **noise** (`cluster -1`, purple color).

---

### **2. High Noise Ratio:**

* Since DBSCAN works based on density, points in low-density regions are labeled as noise (outliers).
* The graph contains many **purple points** (noise), indicating that the model considers a large portion of the data as outliers.

---

### **3. Parameter Settings Might Be Problematic:**

* The current DBSCAN parameters `eps=0.5` and `min_samples=5` might be too strict for this dataset.
* Result: Only one cluster is detected, and the rest are excluded.

- As alternatives, you can try:

* Increasing the `eps` value (e.g., `eps=0.8`, `1.0`, `1.2`)
* Decreasing the `min_samples` value (e.g., to `3` or `4`)

```python
# EM (Gaussian Mixture Model - GMM)
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture

data = pd.read_csv("/Users/mac/Desktop/Mall_Customers.csv")

data_selected = data[['Age', 'Annual Income (k$)']]

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_selected)

gmm = GaussianMixture(n_components=5, random_state=123)  # 5 bileşen ile
gmm_result = gmm.fit_predict(data_scaled)

data['gmm_cluster'] = gmm_result

plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Annual Income (k$)'], c=data['gmm_cluster'], cmap='viridis', s=50)
plt.title('Gaussian Mixture Model Clustering (5 components)')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.colorbar(label='Cluster')
plt.show()
```



### **Dataset: Mall\_Customers.csv**

* Variables used: **Age** and **Annual Income (k\$)**
* Objective: Segment customers based on these two features

---

## **Interpretation of GMM Results**

### 1. **Clustering Successful with Gaussian Mixture Model**

* The model was run with 5 components (clusters), dividing the data into 5 groups.
* Each cluster is based on a **Gaussian distribution (normal distribution)** assumption. This means clusters can be elliptical and can overlap (unlike DBSCAN or KMeans).

---

### 2. **Distinct Distribution Between Clusters:**

* **Cluster 0 (dark purple):**
  Generally **ages 30–45**, **high income (70–120k\$)** customers.

* **Cluster 1 (blue):**

* **Cluster 2 (turquoise-green tones):**

  Lower spending but with growth potential.

* **Cluster 3 (light green):**
  **Young (20–30 years)** but **high income (60–80k\$)** individuals.
  Possibly tech-savvy consumers who follow trends.

* **Cluster 4 (yellow):**
  Can include people of all ages but generally **high income**.
  Spending power is high, though age may vary.
